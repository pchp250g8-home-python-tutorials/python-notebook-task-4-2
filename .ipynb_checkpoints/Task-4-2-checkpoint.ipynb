Партия товара была продана за 864 руб. Прибыль составила 8%. Какова себестоимость товара?

```python
v = 864
ps = 8
p = v * ps /(100 + ps)
s = v - p
print(s)

```

```python
from sympy.abc import x
from sympy import solve
from sympy import Eq

ans = solve(Eq(x + 8 * x / 100, 864), x)
print(ans)
```

```python
from sympy.abc import x, y
from sympy import solve
from sympy import Eq

ans = solve([Eq(x + y, 864), Eq(x / y * 100, 8)], [x, y])
print(ans)
```
